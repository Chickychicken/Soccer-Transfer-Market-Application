```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
df = pd.read_csv('Model_Comparison.csv')
```

```python
df
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Models'], df['Attacker'])
plt.ylabel('RMSE')
plt.title('RMSE of Attackers')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f} million')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Models'], df['Midfielder'])
plt.ylabel('RMSE')
plt.title('RMSE of Midfielders')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f} million')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df['Models'], df['Defender'])
plt.ylabel('RMSE')
plt.title('RMSE of Defenders')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f} million')
plt.gca().yaxis.set_major_formatter(formatter)
plt.show()
```

```python
new_df = pd.DataFrame({
'Model' : ['Attackers (Random Forest)', 'Midfielders (ElasticNet Regression)', 'Defenders (Random Forest)'],
'RMSE': [9949750.28, 8552402.28, 5726260.2]
})
```

```python
new_df
```

```python
for i, value in enumerate(df['Values']):
    plt.text(i, value, str(value), ha='center', va='bottom')
```

```python
plt.figure(figsize = [10, 6])
plt.bar(new_df['Model'], new_df['RMSE'])
plt.title('Results of Best Model From Each Position')
plt.ylabel('RMSE')
formatter = ticker.FuncFormatter(lambda x, pos: f'{x/1e6:.1f} million')
plt.gca().yaxis.set_major_formatter(formatter)
for i, value in enumerate(new_df['RMSE']):
    plt.text(i, value, str(value), ha = 'center')
plt.show()
```
