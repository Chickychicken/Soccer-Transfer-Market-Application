```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# 2023

```python
df = pd.read_csv('transfermarkt_9_leagues_2023.csv')

list = df['Values']
list = [item[:-1] for item in list]

def process_strings(strings):
    cleaned_values = []
    for string in strings:
        string = string[1:]
        if string.endswith('m\xa0'):
            if '.' in string:
                string = string.replace('.', '')
            string = string.replace('m\xa0', '') + '0000'
            cleaned_values.append(int(string))
        elif string.endswith('k\xa0'):
            if '.' in string:
                string = string.replace('.', '')
            string = string.replace('k\xa0', '') + '0000'
            cleaned_values.append(int(string))
        elif string.endswith('m'):
            if '.' in string:
                string = string.replace('.', '')
            string = string.replace('m', '') + '0000'
            cleaned_values.append(int(string))
        elif string.endswith('k'):
            string = string.replace('k', '') + '000'
            cleaned_values.append(int(string))
    return cleaned_values
l = process_strings(list)

final_list = [float(element) for element in l]
df['Values'] = final_list

df = df.sort_values('Values',ascending=False)
df_2023 = df.drop_duplicates()
df_2023.head(10)
```

```python
df_2023.to_csv('2023_playervalues.csv')
```

# 2021

```python
df = pd.read_csv('transfermarkt_9_leagues_2021.csv')

list = df['Values']
list = [item[:-1] for item in list]

def process_strings(strings):
    cleaned_values = []
    for string in strings:
        string = string[1:]
        if string.endswith('m'):
            if '.' in string:
                string = string.replace('.', '')
            string = string.replace('m', '') + '0000'
            cleaned_values.append(int(string))
        elif string.endswith('k'):
            string = string.replace('k', '') + '000'
            cleaned_values.append(int(string))
    return cleaned_values
l = process_strings(list)

final_list = [float(element) for element in l]
df['Values'] = final_list

df = df.sort_values('Values',ascending=False)
df_2021 = df.drop_duplicates()
df_2021 = df_2021.rename(columns = {'Values': 'Previous Values'})
df_2021.head(10)
```

# Finding value difference

```python
difference = pd.merge(df_2023, df_2021, on = 'Players', how = 'inner')
```

```python
difference['dif'] = difference['Values'] - difference['Previous Values']
difference = difference.drop(['Values', 'Previous Values'], axis = 1)
difference = difference.rename(columns = {'dif' : 'Values'})
difference = difference.sort_values('Values',ascending=False)
```
