# Changing Name Order of FBREF Dataset

```python
def process_name(name):
    name_parts = name.split(' ')
    if len(name_parts) >= 3:
        space_index = name.find(' ')
        name = name[:space_index] + '$' + name[space_index+1:]
    return name

# Apply the function to the 'Player' column
FB['Players'] = FB['Players'].apply(process_name)

# Print the updated DataFrame
FB
```

```python
FB['Players'] = FB['Players'].apply(lambda x: ' '.join(x.split()[::-1]))
FB.head(5)
```

```python
def remove_dollar_sign(name):
    return name.replace('$', ' ')

# Apply the function to the 'Player' column where names contain a dollar sign
FB.loc[FB['Players'].str.contains('\$'), 'Players'] = FB.loc[FB['Players'].str.contains('\$'), 'Players'].apply(remove_dollar_sign)

# Print the updated DataFrame
FB.head(5)
```

# Dealing With Duplicated Names

```python
duplicated_names = diff[diff.duplicated(subset='Players', keep=False)]['Players']

for name, count in duplicated_names.value_counts().items():
    if count > 1:
        print(f"Duplicate name: {name}")
        players = diff[diff['Players'] == name]
        for index, player in players.iterrows():
            print(f"    {player['Players']} - Age: {player['Age']}, Value: {player['Values']}")
```

```python
diff.drop_duplicates(subset='Players', keep=False, inplace=True)
```

```python
duplicated_names = FB[FB.duplicated(subset='Players', keep=False)]['Players']

for name, count in duplicated_names.value_counts().items():
    if count > 1:
        print(f"Duplicate name: {name}")
        players = FB[FB['Players'] == name]
        for index, player in players.iterrows():
            print(f"    {player['Players']} - Age: {player['Age']}, Matches Played: {player['Matches Played']}")
```
