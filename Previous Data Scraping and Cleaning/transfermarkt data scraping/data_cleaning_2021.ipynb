```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Premier League

```python
PL = pd.read_csv('PL2021.csv')

list = PL['Values']
list = [item[:-1] for item in list]

def process_strings(strings):
    cleaned_values = []
    for string in strings:
        string = string[1:]
        if string.endswith('m'):
            if '.' in string:
                string = string.replace('.', '')
            string = string.replace('m', '') + '0000'
            cleaned_values.append(int(string))
        elif string.endswith('k'):
            string = string.replace('k', '') + '000'
            cleaned_values.append(int(string))
    return cleaned_values
l = process_strings(list)

final_list = [float(element) for element in l]
PL['Values'] = final_list

PL = PL.sort_values('Values',ascending=False)
PL = PL.drop_duplicates()
PL.head(10)
```
