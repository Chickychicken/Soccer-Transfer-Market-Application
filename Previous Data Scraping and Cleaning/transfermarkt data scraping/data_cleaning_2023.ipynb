```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Ligue 1 value converting

```python
ligue = pd.read_csv('transfermarkt2.csv')
ligue.drop_duplicates(subset=['Player'], keep='first', inplace = True)
ligue.head(25)
```

```python
list = ligue['Value']
list = [item[:-1] for item in list]
```

```python
def process_strings(strings):
    processed_strings = []
    for string in strings:
        if '.' in string:
            string = string.replace('.', '') + '0'
        processed_strings.append(string)
    return processed_strings
l = process_strings(list)

final_list = []
for a in l:
    final_list.append(a + '000.0')

final_list = [float(element) for element in final_list]
ligue['Value'] = final_list

ligue = ligue.sort_values('Value',ascending=False)
ligue.head(10)
```

# Merging all dataframes together

```python
df = pd.read_csv('transfermarkt.csv')
final_df = pd.concat([df, ligue], ignore_index=True)
final_df.sort_values('Value',ascending=False, inplace = True)
final_df = final_df.reset_index(drop = True)
final_df.head(10)
```

# Adding league column

```python
#Grouping Clubs by individual Leagues and making a separate "League" column.
England = ['Manchester City','Arsenal FC','Chelsea FC','Manchester United','Liverpool FC','Tottenham Hotspur','Newcastle United','Aston Villa','Brighton & Hove Albion','West Ham United','Leicester City','Wolverhampton Wanderers','Southampton FC','Brentford FC','Nottingham Forest','Everton FC','Leeds United','Crystal Palace','AFC Bournemouth','Fulham FC']
Spain = ['FC Barcelona','Real Madrid','Atlético de Madrid','Real Sociedad','Villarreal CF','Real Betis Balompié','Athletic Bilbao','Valencia CF','Sevilla FC','CA Osasuna','Celta de Vigo','Girona FC','Getafe CF','RCD Espanyol Barcelona','RCD Mallorca','Rayo Vallecano','UD Almería','Real Valladolid CF','Elche CF','Cádiz CF']
Italy = ['SSC Napoli','Inter Milan','AC Milan','Juventus FC','Atalanta BC','AS Roma','SS Lazio','ACF Fiorentina','Torino FC','Udinese Calcio','Benevento Calcio','US Sassuolo','Bologna FC 1909','AC Monza','US Salernitana 1919','FC Empoli','US Lecce','Spezia Calcio','Hellas Verona','US Cremonese','UC Sampdoria']
Germany = ['Bayern Munich','Borussia Dortmund','RB Leipzig','Bayer 04 Leverkusen','Eintracht Frankfurt','Borussia Mönchengladbach','VfL Wolfsburg','SC Freiburg','1.FC Union Berlin','TSG 1899 Hoffenheim','1.FSV Mainz 05','VfB Stuttgart','FC Augsburg','1.FC Köln','SV Werder Bremen','Hertha BSC','FC Schalke 04','VfL Bochum']
France = ['Paris Saint-Germain','Stade Rennais FC','AS Monaco','Olympique Marseille','Olympique Lyon','LOSC Lille','OGC Nice','RC Lens','Montpellier HSC','FC Nantes','Stade Reims','RC Strasbourg Alsace','Stade Rennais FC','Dijon FCO','RC Strasbourg Alsace','FC Lorient','FC Toulouse','ESTAC Troyes','Stade Brestois 29','AJ Auxerre','Angers SCO','Clermont Foot 63','AC Ajaccio']

LeagueList = []
for a in final_df['Club']:
    if a in England:
        LeagueList.append('Premier League')
    elif a in Spain:
        LeagueList.append('La Liga')
    elif a in Italy:
        LeagueList.append('Serie A')
    elif a in Germany:
        LeagueList.append('Bundesliga')
    elif a in France:
        LeagueList.append('Ligue 1')
    else:
        LeagueList.append("UNKNOWN")
        
final_df['League'] = LeagueList
```
