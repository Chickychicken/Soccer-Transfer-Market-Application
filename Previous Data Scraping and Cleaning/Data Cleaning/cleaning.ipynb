```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
FBREF = pd.read_csv("final_FBREF_data.csv")
tran = pd.read_csv("final_transfermarkt_data.csv")
tran_2021 = pd.read_csv("2021values.csv")
```

```python
FBREF.drop('Position', axis = 1, inplace = True)
```

# Replace special letters

```python
from unidecode import unidecode
```

```python
FBREF['Player'] = FBREF['Player'].apply(unidecode)
tran['Player'] = tran['Player'].apply(unidecode)
tran_2021['Players'] = tran_2021['Players'].apply(unidecode)
```

```python
FBREF.head(5)
```

```python
tran.head(5)
```

```python
tran_2021 = tran_2021.rename(columns={'Players': 'Player'})
tran_2021
```

# Re ordering names from FBREF

```python
def process_name(name):
    name_parts = name.split(' ')
    if len(name_parts) >= 3:
        space_index = name.find(' ')
        name = name[:space_index] + '$' + name[space_index+1:]
    return name

# Apply the function to the 'Player' column
FBREF['Player'] = FBREF['Player'].apply(process_name)

# Print the updated DataFrame
FBREF.head(4)
```

```python
FBREF['Player'] = FBREF['Player'].apply(lambda x: ' '.join(x.split()[::-1]))
FBREF.head(5)
```

```python
def remove_dollar_sign(name):
    return name.replace('$', ' ')

# Apply the function to the 'Player' column where names contain a dollar sign
FBREF.loc[FBREF['Player'].str.contains('\$'), 'Player'] = FBREF.loc[FBREF['Player'].str.contains('\$'), 'Player'].apply(remove_dollar_sign)

# Print the updated DataFrame
FBREF.head(5)
```

# Dealing with Duplicated Names

```python
duplicates = tran[tran.duplicated(subset='Player', keep=False)]
print(duplicates)
```

```python
duplicated_names = FBREF[FBREF.duplicated(subset='Player', keep=False)]['Player']

for name, count in duplicated_names.value_counts().items():
    if count > 1:
        print(f"Duplicate name: {name}")
        players = FBREF[FBREF['Player'] == name]
        for index, player in players.iterrows():
            print(f"    {player['Player']} - Age: {player['Age']}, Matches Played: {player['Matches Played']}")
```

```python
FBREF.sort_values('Age', ascending=False, inplace=True)
tran.sort_values('Age', ascending=False, inplace=True)

# Drop duplicates based on 'Player' column, keeping the first occurrence (older age)
FBREF.drop_duplicates(subset='Player', keep='first', inplace=True)
tran.drop_duplicates(subset='Player', keep='first', inplace=True)
```

```python
duplicates = tran[tran.duplicated(subset='Player', keep=False)]
print(duplicates)
```

```python
duplicates = FBREF[FBREF.duplicated(subset='Player', keep=False)]
print(duplicates)
```

```python
tran.drop('Age', axis = 1, inplace = True)
```

# Merging data

```python
merged_df = pd.merge(FBREF, tran, on=['Player'], how='left')
```

```python
merged_df.dropna(subset=merged_df.columns[-5:], inplace=True)
merged_df = merged_df.sort_values('Value', ascending=False)
merged_df = merged_df.reset_index()
merged_df.fillna(0, inplace=True)
```

```python
columns = merged_df.columns.tolist()
columns[-1], columns[-2] = columns[-2], columns[-1]
df = merged_df[columns]
```

```python
df
```

```python
df.to_csv('dataset.csv', index=False)
```

# Value Differnece Dataset

```python
merged_df = pd.merge(tran_2021, tran, on=['Player'], how='left')
```

```python
merged_df
```

```python
merged_df.dropna(subset=merged_df.columns[-5:], inplace=True)
merged_df = merged_df.sort_values('Value', ascending=False)
merged_df = merged_df.reset_index()
merged_df.fillna(0, inplace=True)
```

```python
merged_df['Cost'] = merged_df['Value'] - merged_df['Values']
merged_df = merged_df.drop(['Value', 'Values'], axis=1)
merged_df = merged_df.rename(columns={'Cost': 'Value'})
merged_df = merged_df.sort_values('Value', ascending=False)
```

```python
columns = merged_df.columns.tolist()
columns[-1], columns[-2] = columns[-2], columns[-1]
df = merged_df[columns]
df = df.iloc[:, 1:]
```

```python
df.head(25)
```

```python
merged_df = pd.merge(FBREF, df, on=['Player'], how='left')
```

```python
merged_df.dropna(subset=merged_df.columns[-5:], inplace=True)
merged_df = merged_df.sort_values('Value', ascending=False)
merged_df = merged_df.reset_index()
merged_df.fillna(0, inplace=True)
```

```python
columns = merged_df.columns.tolist()
columns[-1], columns[-2] = columns[-2], columns[-1]
df = merged_df[columns]
```
